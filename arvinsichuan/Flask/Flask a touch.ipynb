# Flask a touch

Flask is one of the popular web development framework using python.  

It is very easy to start a web application using Flask

```python
from flask import Flask

app = Flask(__name__)


@app.route('/')
def hello_world():
    return 'Hello World!'


if __name__ == '__main__':
    app.run()
```
Above is a quick demo of how it works.

## Templates
Templates is a kind of static files that will be rendered during the process of requesting a page. By using something like a template engine, Flask will assemble the contents as you wish.

- Directory for templates
```
├ templates
├─ hello.html
└ app.py
```
The `app.py` is the main entrance for the application while the templates is a folder includes the `.html` files.
