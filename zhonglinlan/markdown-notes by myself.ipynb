# markdown常用语法自整理
1. 使用#加空格表示标题，多个#表示多级标题，例如：
# 标题1
## 标题2
### 标题3

2. 使用+，-，*加空格表示无序列表，例如：
 + 无序列表项1
 - 无序列表项2
 * 无序列表项3

3. 使用数字加英文句点.来表示有序列表，若缩进与该列表项相同，即使使用无序列表标识符或者乱序，会默认生成有序列表项，例如：
 1. 有序列表项1
 + 有序列表项2
 6. 有序列表项3
 <br>注：使用混合列表时，具有相同缩进的项为同一列表的项<br>

4. 使用&lt;br&gt;   &lt;br&gt;表示换行，或者两行之间隔一行表示换行。例如：
  <br>这是一行<br>
  
  这是另一行
5. 在一行中用三个以上的*、-、_来建立一个分隔线，行内不能有其他东西。也可以在符号间插入空格,例如：

 分割线
 - - - - - - 
 再来一根
 * * * * * * 
 最后一根
 _ _ _ _ _ _

6. 两个*或-代表加粗，一个*或-代表斜体，两个~代表删除，例如：

**加粗文本** 或者 __加粗文本__

*斜体文本*  或者_斜体文本_

~~删除文本~~

7. 表格对齐格式居左：:----居中：:----:或----- 居右：----，例如：


 |标题|标题|标题|
 |:---|:---:|---:|
 |居左测试文本|居中测试文本|居右测试文本|
 |居左测试文本1|居中测试文本2|居右测试文本3|
 |居左测试文本11|居中测试文本22|居右测试文本33|
 |居左测试文本111|居中测试文本222|居右测试文本333|
  
  <br>注：在列表中生成表格，要空两行才有效<br>

                               <br>看见了吗？<br>
